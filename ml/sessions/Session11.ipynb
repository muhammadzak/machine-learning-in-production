```python
%matplotlib inline
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creatig dataset
x_values = np.array(range(10)).reshape(-1, 1)
y_values = np.array([k**.5 for k in range(10)])
print x_values
print y_values
```

```python
#train model on data
dt = DecisionTreeRegressor()
dt.fit(x_values, y_values)
```

```python
#predict one values
y_predicted = dt.predict(2)
print y_predicted
```

```python
#visualize actual data
plt.scatter(x_values, y_values)
plt.plot(x_values, y_values)
plt.show()
```

```python
#visualize results
plt.scatter(x_values, y_values)
plt.plot(x_values, dt.predict(x_values))
plt.show()
```
